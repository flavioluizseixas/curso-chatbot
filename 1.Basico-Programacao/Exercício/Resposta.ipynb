```python
# Sistema de Controle de Tarefas com Prioridade

# Dicionário para armazenar as tarefas organizadas por prioridade
tarefas = {
    "alta": [],
    "media": [],
    "baixa": []
}
```

```python
def adicionar_tarefa():
    tarefa = input("Digite o nome da tarefa: ")
    prioridade = input("Escolha a prioridade (alta, media, baixa): ").lower()

    if prioridade in tarefas:
        tarefas[prioridade].append(tarefa)
        print(f"Tarefa '{tarefa}' adicionada com prioridade {prioridade}.")
    else:
        print("Prioridade inválida. Tarefa não adicionada.")
```

```python
def listar_tarefas():
    print("\nTarefas Pendentes:")
    for prioridade, lista in tarefas.items():
        print(f"\nPrioridade {prioridade.capitalize()}:")
        if not lista:
            print("  Nenhuma tarefa.")
        else:
            for idx, tarefa in enumerate(lista, 1):
                print(f"  {idx}. {tarefa}")
```

```python
def concluir_tarefa():
    prioridade = input("Escolha a prioridade da tarefa que deseja marcar como concluída (alta, media, baixa): ").lower()

    if prioridade in tarefas and tarefas[prioridade]:
        listar_tarefas()
        num_tarefa = int(input(f"Digite o número da tarefa que deseja concluir (Prioridade {prioridade}): "))

        if 1 <= num_tarefa <= len(tarefas[prioridade]):
            tarefa_concluida = tarefas[prioridade].pop(num_tarefa - 1)
            print(f"Tarefa '{tarefa_concluida}' concluída e removida.")
        else:
            print("Número de tarefa inválido.")
    else:
        print("Prioridade inválida ou não há tarefas nesta prioridade.")
```

```python
def menu():
    while True:
        print("\nMenu:")
        print("1. Adicionar nova tarefa")
        print("2. Listar tarefas pendentes")
        print("3. Marcar tarefa como concluída")
        print("4. Sair")

        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            adicionar_tarefa()
        elif escolha == "2":
            listar_tarefas()
        elif escolha == "3":
            concluir_tarefa()
        elif escolha == "4":
            print("Saindo do programa. Até mais!")
            break
        else:
            print("Opção inválida. Tente novamente.")
```

```python
# Inicia o programa
menu()
```
