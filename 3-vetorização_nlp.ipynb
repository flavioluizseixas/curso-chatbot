Quando a Tokenização Não é Necessária?
Se você estiver utilizando modelos como TF-IDF ou BERT através de frameworks como scikit-learn ou transformers, a tokenização pode ser feita automaticamente, sem que você precise se preocupar com ela.

```python
# Texto de exemplo
texto = "Olá! Eu sou um chatbot. Como posso ajudá-lo hoje?"
```

```python
#tokenização
tokens = word_tokenize(texto)
```

```python
# Vetorização com TF-IDF
vectorizer = TfidfVectorizer()
vetor_tfidf = vectorizer.fit_transform(tokens)

print("Vetor TF-IDF:")
print(vetor_tfidf.toarray())
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Carregar o modelo BERTimbau e o tokenizer
tokenizer = AutoTokenizer.from_pretrained("neuralmind/bert-base-portuguese-cased")
model = AutoModel.from_pretrained("neuralmind/bert-base-portuguese-cased")

# Tokenização e vetorização feitas automaticamente
tokens = tokenizer(texto, return_tensors="pt", padding=True, truncation=True)
with torch.no_grad():
    embeddings = model(**tokens)

# Extração dos embeddings
vetor_bert = embeddings.last_hidden_state[:, 0, :].numpy()

print("Vetor BERT:")
print(vetor_bert)
```
